```python
## Independent and Dependent features

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
 ## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_test
```

```python
## Standardize the dataset
## Different features have different ranges (e.g., crime rate vs. tax rate).
## Scaling makes sure all features are on the same scale, so no one feature dominates.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

# Model training

```python
## This line teaches the model how the features relate to the house price.
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
## # weights for each feature
print(regression.coef_)
```

```python
# bias term
print(regression.intercept_)
```

```python
## on which parameter this model has been trained
regression.get_params()
```

```python
### Prediction with Test Data
## Now your model predicts the prices for the houses in the test set.

reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

## Assumptions

```python
import matplotlib.pyplot as plt

# Example condition: points where prediction < actual are red, others blue
colors = ['red' if pred < actual else 'blue' for actual, pred in zip(y_test, reg_pred)]

plt.scatter(y_test, reg_pred, color=colors)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values (Red & Blue)")
plt.show()
```

```python
## It shows how wrong your model was.
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals

sns.displot(residuals,kind="kde")
plt.show()
```

```python
## Scatter plot with respect to prediction and residuals
## uniform distribution
plt.scatter(reg_pred,residuals)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred)) # average error in price
print(mean_squared_error(y_test,reg_pred)) # squares the errors
print(np.sqrt(mean_squared_error(y_test,reg_pred))) # RMSE (in price units)
```

## R square and adjusted R square

Formula

R^2=1- SSR/SST

R^2= coefficient of determination SSR = sum of squares of residuals SST = total sum of sqaures

 

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,reg_pred)
print(score)
```

Adjusted R2 = 1- [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
#display adjusted R-sqaured
1- (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
